```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}

for product in products:
    quantity_of_product = int(input(f"Please enter the quantity of {product}:"))
    inventory[product] = quantity_of_product

customer_orders = set()
```

```python
while True:
    order = input("Enter the name of the product to order: ").strip()

    if order not in products:
        print(f"{order} is not available. Please choose a valid product.")
        continue
        
    if order in customer_orders:
        print(f"{order} ir already in the order.")
    else:
        customer_orders.add(order)
        print(f"{order} added to your order.")

    another = input("Do you want to add another product? (yes/no): ").strip().lower()
    if another != "yes":
        break
```

```python
customer_orders

for product in customer_orders:
    if inventory[product] > 0:
        inventory[product] -= 1
    else:
        print(f"Sorry, {product} is out of stock!")

print(f"Customer Order:", customer_orders)
print(f"Updated Inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
```
